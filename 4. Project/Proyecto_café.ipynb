# Proyecto "Predicción de preferencia por tueste de café de acuerdo con variables socioeconómicas"

## Pregunta de investigación: 
**¿Qué características sociodemográficas predicen de manera más efectivas la preferencia de tipo de café en los consumidores, y cómo puede esta información mejorar las elecciones de bebidas en cafeterías de especialidad?**

## Decripción del dataset
Prueba de sabor de café, realizada por 4027 consumidores en EE.UU. Se cataron cuatro cafés diferentes con kits Cometeer (cápsulas de extracto de café) para obtener una experiencia de sabor prácticamente idéntica. La cata y las calificaciones se realizaron a ciegas. Se recopilaron los resultados de la evaluación personal, así como de los datos socioeconómicos, a través de cuestionarios.

Por lo tanto, se proponen dos clases similares: <br>
- Clase 1: **'What roast level of coffee do you prefer?'**
- Clase 2: **'Before today's tasting, which of the following best described what kind of coffee you like?'**

```python
def calculate_confusion_matrix(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    labels = np.unique(y_true)
    confusion_matrix = np.zeros((len(labels), len(labels)))
    for i in range(y_true.shape[0]):
        true_index = np.where(labels == y_true[i])[0]
        pred_index = np.where(labels == y_pred[i])[0]
        confusion_matrix[true_index,pred_index] += 1

    return confusion_matrix
```

```python
# Evaluación de modelo
def evaluate_models(models, X_train, y_train):    
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    fig, axs = plt.subplots(4, 3, figsize=(16, 9)) 
    axs = axs.flatten()
    i = 0

    results_data = {'model': [], 'accuracy': [], 'precision': [], 'recall': [], 'f1': []}

    for name, model in models.items():
        acc_scores = []
        for train_index, test_index in skf.split(X_train, y_train):
            # Particiones para training y testing
            X_train_fold, X_test_fold = X_train.iloc[train_index], X_train.iloc[test_index]
            y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]

            # Entrenamiento y evaluación del modelo
            model.fit(X_train_fold, y_train_fold)
            y_predict = model.predict(X_test_fold)

            # Calcular métricas
            acc = accuracy_score(y_test_fold, y_predict)
            acc_scores.append(acc)

            # Guardar métricas en el diccionario
            results_data['model'].append(name)
            results_data['accuracy'].append(acc)
            results_data['precision'].append(precision_score(y_test_fold, y_predict, average='macro'))
            results_data['recall'].append(recall_score(y_test_fold, y_predict, average='macro'))
            results_data['f1'].append(f1_score(y_test_fold, y_predict, average='macro'))

        # Matriz de confusión
        cm = calculate_confusion_matrix(y_test_fold, y_predict)
        sns.heatmap(cm, annot=True, fmt='.0f', ax=axs[i], cmap='Blues')
        axs[i].set_title(name)
        axs[i].set_xlabel('Predicted')
        axs[i].set_ylabel('True')
        i += 1  # Incrementar el índice para acceder al siguiente subplot

        avg_acc = sum(acc_scores) / len(acc_scores)
        print(f"{name}: Avg accuracy {avg_acc:.4f}")
        print(classification_report(y_test_fold, y_predict))

    # Ocultar gráficos sobrantes
    for j in range(i, len(axs)):
        fig.delaxes(axs[j])

    plt.tight_layout()
    plt.show()

    return results_data
```

```python
# Gráfico de métricas
def plot_models(results_data):
    fig, axs = plt.subplots(2, 2, figsize=(16, 12))  # Tamaño ajustado para evitar solapamientos

    # Lista de métricas
    metrics = ['accuracy', 'precision', 'recall', 'f1']
    titles = ['Accuracy', 'Precision', 'Recall', 'F1-score']

    # Crear cada gráfico
    for idx, (metric, title) in enumerate(zip(metrics, titles)):
        row, col = divmod(idx, 2) 
        sns.boxplot(data=results_data, x='model', y=metric, palette='rocket', hue='model', ax=axs[row, col])
        axs[row, col].set_title(title)
        axs[row, col].tick_params(axis='x', rotation=45)  # Rotar etiquetas en el eje x
        if axs[row, col].legend_:
            axs[row, col].legend_.remove()
    plt.tight_layout()

    # Intentar generar una leyenda global si hay handles y labels
    handles, labels = axs[0, 0].get_legend_handles_labels()
    if handles and labels:
        fig.legend(handles, labels, loc='upper center', ncol=len(labels), bbox_to_anchor=(0.5, 1.05))
    
    plt.show()
```

```python
# Optimiza los hiperparámetros de un modelo utilizando RandomizedSearchCV o GridSearchCV.
def optimize_model(models, params_models, X, y, search_type='random'):
    common_params = {
        'scoring': 'f1',
        'cv': 5,
        'verbose': 0,
        'n_jobs': -1
    }

    results = {}

    for model_name, model in models.items():
        print(f"Optimizing {model_name}...")
        
        # Determinar el tipo de búsqueda
        if search_type == 'random':
            search = RandomizedSearchCV(
                estimator=model,
                param_distributions=params_models[model_name], 
                n_iter=50,
                random_state=42,
                **common_params
            )
        elif search_type == 'grid':
            search = GridSearchCV(
                estimator=model,
                param_grid=params_models[model_name], 
                **common_params
            )
        else:
            raise ValueError("El parámetro 'search_type' debe ser 'grid' o 'random'.")

        search.fit(X, y)       
        results[model_name] = {
            'best_params': search.best_params_,
            'best_score': search.best_score_,
            'best_model': search.best_estimator_}

    return results
```
