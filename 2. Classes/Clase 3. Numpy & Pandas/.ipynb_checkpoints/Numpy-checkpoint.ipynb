# NumPy

The NumPy library is the core library for scientific computing in Python. It provides a high-performance multidimensional array
 object, and tools for working with these arrays

 **Use the following import convention**


```python
import numpy as np
```

### NumPy Arrays

```python
from IPython.display import Image
Image(filename='images/1. NumPy Arrays.png') 
```

## Creating Arrays

```python
a = np.array([1,2,3])
```

```python
b = np.array([(1.5,2,3), (4,5,6)], dtype = float)
```

```python
c = np.array([[(1.5,2,3), (4,5,6)],[(3,2,1), (4,5,6)]], dtype = float)
```

```python
np.zeros((3,4)) #Create an array of zeros
```

```python
np.ones((2,3,4),dtype=np.int16) #Create an array of ones
```

```python
d = np.arange(10,25,5) #Create an array of evenly spaced values (step value)
```

```python
np.linspace(0,2,9) #Create an array of evenly spaced values (number of samples)
```

```python
e = np.full((2,2),7) #Create a constant array
```

```python
f = np.eye(2) #Create a 2X2 identity matrix
```

```python
np.random.random((2,2)) #Create an array with random values
```

```python
np.empty((3,2)) #Create an empty array
```

## Array Mathematics

### Arithmetic Operations

```python
a - b #Subtraction
```

```python
np.subtract(a,b) #Subtraction numpy version
```

```python
b + a #Addition
```

```python
np.add(b,a) #Addition numpy version
```

```python
a / b # Division
```

```python
np.divide(a,b) #Division numpy version
```

```python
a * b #Multiplication
```

```python
np.multiply(a,b) #Multiplication numpy version
```

```python
np.exp(b) #Exponentiation
```

```python
np.sqrt(b) #Square root
```

```python
np.sin(a) #Print sines of an array
```

```python
np.cos(b) #Element-wise cosine
```

```python
np.log(a) #Element-wise natural logarithm
```

```python
e.dot(f) #Dot product
```

### Comparison

```python
a == b #Element-wise comparison
```

```python
a < 2 #Element-wise comparison
```

```python
np.array_equal(a, b) #Array-wise comparison
```

### Aggregate Functions

```python
a.sum() #Array-wise sum
```

```python
a.min() #Array-wise minimum value
```

```python
b.max(axis=0) #Maximum value of an array row
```

```python
b.cumsum(axis=1) #Cumulative sum of the elements
```

```python
a.mean() #Mean
```

```python
np.median(b) #Median
```

```python
np.corrcoef(a) #Correlation coefficient
```

```python
np.std(b) #Standard deviation
```

## Copying Arrays

```python
h = a.view() #Create a view of the array with the same data
```

```python
np.copy(a) #Create a copy of the array
```

```python
h = a.copy() #Create a deep copy of the array
```
